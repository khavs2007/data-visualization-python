```python
import pandas as pd
import matplotlib.pyplot as plt


data = {

    'Month': list(range(1, 13)),
    'Sales_Product_A': [100, 120, 150, 130, 180, 200, 250, 220, 280, 300, 320, 350],
    'Sales_Product_B': [150, 160, 140, 180, 190, 220, 200, 240, 250, 270, 290, 310]
}
df = pd.DataFrame(data)


plt.figure(figsize=(10, 6))


plt.plot(df['Month'], df['Sales_Product_A'], label='Product A', marker='o', linestyle='-', color='blue')


plt.plot(df['Month'], df['Sales_Product_B'], label='Product B', marker='x', linestyle='--', color='orange')


plt.title('Monthly Sales Comparison (Product A vs. Product B)')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.legend(title='Product')
plt.grid(True)
plt.xticks(df['Month'])
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


my_scores = [
    65, 78, 85, 92, 70, 75, 88, 60, 95, 80,
    72, 81, 89, 77, 91, 68, 73, 84, 86, 76,
    82, 90, 79, 71, 87, 69, 74, 83, 94, 66
]


data = {
    'Scores': my_scores
}
df = pd.DataFrame(data)

COLUMN_TO_PLOT = 'Scores'




sns.set_style("whitegrid")


fig, (ax_box, ax_hist) = plt.subplots(
    2,
    1,
    sharex=True,
    gridspec_kw={"height_ratios": (0.15, 0.85)}
)

sns.boxplot(
    x=df[COLUMN_TO_PLOT],
    ax=ax_box,
    color="lightcoral"
)

ax_box.set(xlabel='')

sns.histplot(
    data=df,
    x=COLUMN_TO_PLOT,
    ax=ax_hist,
    kde=True,
    bins=10,
    color="firebrick"
)


plt.suptitle(f"Distribution of {COLUMN_TO_PLOT} with Boxplot and Histogram")
plt.show()
```

```python
#E-commerce / User Experience
import pandas as pd
import matplotlib.pyplot as plt
import math
num_days = 50
dates = pd.to_datetime(pd.date_range(start='2025-01-01', periods=num_days, freq='D'))


base_trend = [100 + (50 / num_days) * i for i in range(num_days)]


amplitude = 5
frequency = 0.5
values = [
    (base_trend[i] + amplitude * math.sin(i * frequency)).__round__(2)
    for i in range(num_days)
]

df = pd.DataFrame({
    'Date': dates,
    'Daily_Sales': values
})


plt.figure(figsize=(10, 6))


plt.plot(df['Date'], df['Daily_Sales'], color='#0077B6', linestyle='-', linewidth=2, marker='o', markersize=4, label='Daily Sales (Trend + Cycle)')


df['7_Day_Avg'] = df['Daily_Sales'].rolling(window=7).mean()
plt.plot(df['Date'], df['7_Day_Avg'], color='#E24A33', linestyle='--', linewidth=2, label='7-Day Moving Average')



plt.title('Simulated E-commerce Daily Sales Over Time (Deterministic Cycle)', fontsize=16, fontweight='bold')
plt.xlabel('Date', fontsize=12)
plt.ylabel('Sales (in $)', fontsize=12)
plt.legend(loc='upper left')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

plt.show()
```

```python
import matplotlib.pyplot as plt


manual_deployment_hours = 4.0
feature_development_hours = 3.0
bug_fixing_hours = 0.5
meetings_and_admin_hours = 0.5
total_hours = 8.0


activities = [
    'Manual Deployment (Problem)',
    'Feature Development (Core Work)',
    'Bug Fixing',
    'Meetings/Admin'
]
hours = [
    manual_deployment_hours,
    feature_development_hours,
    bug_fixing_hours,
    meetings_and_admin_hours
]


colors = ['#E24A33', '#348AA7', '#6A994E', '#FFC300']


plt.figure(figsize=(8, 8))


explode_settings = (0.1, 0, 0, 0)

plt.pie(
    hours,
    labels=activities,
    colors=colors,

    autopct='%1.1f%%',
    startangle=90,
    shadow=True,
    explode=explode_settings,
    wedgeprops={'edgecolor': 'black', 'linewidth': 1}
)

plt.title(
    'Developer Time Allocation: Impact of Manual Deployment (8-Hour Day)',
    fontsize=15,
    fontweight='bold',
    pad=20
)

plt.tight_layout()
plt.show()
```
